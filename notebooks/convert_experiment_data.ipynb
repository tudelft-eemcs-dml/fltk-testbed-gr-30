```python
path = "/home/jeroen/Documents/CSE/MSc/year/1/Q4/CS4290/repo/fltk-testbed-gr-30/charts/extractor/fig8/output_0.0/"

def mapper(x: str) -> np.array:
    st = ','.join(x.split())
    return np.genfromtxt(StringIO(st[1:-2]), delimiter=',')

data = pd.concat([pd.read_csv(client, converters={'class_precision': mapper, 'class_recall': mapper}) for client in Path(path).rglob('*.csv')])
```
